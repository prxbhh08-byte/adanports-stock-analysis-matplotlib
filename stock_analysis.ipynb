This notebook contains 50 visual analysis plots for time-series EDA.

```python
# Assumption:
# DataFrame name = a
# a['Date'] is already converted to datetime and sorted
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
a = pd.read_csv("ADANIPORTS.csv")
a
```

```python
# 1 Closing Price Trend
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['Close'])
plt.title("Closing Price Trend")
plt.show()
```

```python
# 2 Open Price Trend
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['Open'])
plt.title("Open Price Trend")
plt.show()
```

```python
# 3 High Price Trend
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['High'])
plt.title("High Price Trend")
plt.show()
```

```python
# 4 Low Price Trend
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['Low'])
plt.title("Low Price Trend")
plt.show()
```

```python
# 5 High vs Low
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['High'], label='High')
plt.plot(a['Date'], a['Low'], label='Low')
plt.legend()
plt.title("High vs Low")
plt.show()
```

```python
# 6 Daily Return
a['Daily Return'] = a['Close'].pct_change()
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['Daily Return'])
plt.title("Daily Return")
plt.show()
```

```python
# 7 Return Histogram
plt.figure(figsize=(8,4))
plt.hist(a['Daily Return'].dropna(), bins=30)
plt.title("Return Distribution")
plt.show()
```

```python
# 8 Volume Trend
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['Volume'])
plt.title("Volume Trend")
plt.show()
```

```python
# 9 High-Low Spread
a['HL_Spread'] = a['High'] - a['Low']
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['HL_Spread'])
plt.title("High-Low Spread")
plt.show()
```

```python
# 10 Cumulative Return
a['Cumulative Return'] = (1 + a['Daily Return']).cumprod()
plt.figure(figsize=(10,5))
plt.plot(a['Date'], a['Cumulative Return'])
plt.title("Cumulative Return")
plt.show()
```
