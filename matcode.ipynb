```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv("ADANIPORTS.csv")
a
```

```python
a.head()
```

```python
a.describe()
```

```python
a.info()    
```

```python
# Convert 'Date' column to datetime format
a['Date'] = pd.to_datetime(a['Date'])

# Sort data chronologically
a = a.sort_values('Date')

a.head()
```

```python
# Plot closing price over time
plt.plot(a['Date'], a['Close'])

# Add title and labels
plt.title("ADANIPORTS Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")

# Rotate dates for better visibility
plt.xticks(rotation=45)

# Display the graph
plt.show()
```

```python
# Plot trading volume as bar chart
plt.bar(a['Date'], a['Volume'])

# Add title and labels
plt.title("ADANIPORTS Volume")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate 20-day moving average of closing price
a['MA20'] = a['Close'].rolling(20).mean()

# Plot closing price and moving average together
plt.plot(a['Date'], a['Close'])
plt.plot(a['Date'], a['MA20'])

plt.title("Closing Price with Moving Average")
plt.show()
```

```python
# Calculate daily percentage return
a['Return'] = a['Close'].pct_change()

# Plot daily return
plt.plot(a['Date'], a['Return'])

plt.title("Daily Return")
plt.show()
```

```python
# Find highest and lowest closing price
print("Highest Close:", a['Close'].max())
print("Lowest Close:", a['Close'].min())
```

```python
# Calculate average closing price
print("Average Close:", a['Close'].mean())
```

```python
# Total volume traded
print("Total Volume:", a['Volume'].sum())
```

```python
# Create new column for daily range
a['Range'] = a['High'] - a['Low']

# Plot daily range
plt.plot(a['Date'], a['Range'])
plt.title("Daily Price Range")
plt.show()

```

```python
# Plot histogram of closing prices
plt.hist(a['Close'], bins=20)
plt.title("Closing Price Distribution")
plt.show()
```
